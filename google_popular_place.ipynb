```python
import populartimes
from datetime import datetime
import googlemaps
from motionless import DecoratedMap, LatLonMarker
import operator
```

```python
# geolocationAPI, get current lat,lng
gmaps = googlemaps.Client(key="your_key")
current_position = gmaps.geolocate()
latlon = current_position["location"]
```

```python
# make 500metre grid by lat,lon (in Japan)
lat_width = 0.0090133729745762/2 #500m
lon_width =  0.010966404715491394/2 #500m

#specify the grid
lat_start = latlon["lat"] - lat_width 
lat_end = latlon["lat"] + lat_width
lon_start = latlon["lng"] - lon_width
lon_end = latlon["lng"] + lon_width

res = populartimes.get("your_key", ["bar"], (lat_start, lon_start), (lat_end, lon_end),n_threads=20, radius=100  ,all_places=False) #query=bar
```

```python
# current weekday and hour
weekday = datetime.now().weekday()
current_hour = datetime.now().hour

names, latlon, currnet_normal_populality= [],[],[]

#get place's name, latlon, popularity
for i in range(len(res)):
    names.append(res[i]["name"])
    latlon.append(res[i]["coordinates"])
    currnet_normal_populality.append(res[i]["populartimes"][weekday]["data"][current_hour]    )
```

```python
# sort by popularity
dicts = {key: value for (key, value) in zip(names, currnet_normal_populality)}
sorted_x = sorted(dicts.items(), key=operator.itemgetter(1))
sorted_x
```

```python
from gmplot import gmplot

gmap = gmplot.GoogleMapPlotter(current_position["location"]["lat"], current_position["location"]["lng"], 13)

def color_logic(popularity):
    if popularity >=80:
        color = "red"
    elif popularity >=60:
        color = "orange"
    elif popularity >=40:
        color = "yellow"
    elif popularity >=20:
        color = "green"
    else:
        color = "blue"
    return color

for n,l,c in zip(names, latlon, currnet_normal_populality):
    # Marker
    _lat, _lon = l["lat"], l["lng"]
    ttl = str(n)+" :popularity: "+str(c)
    gmap.marker(lat = _lat, lng = _lon, color = color_logic(c), title = ttl)

# Open "my_map.html" file, created on the same directory
```
