```python
# 周辺情報を引っ張ってくる
# 現在空いてて、飲食店縛り
# 結局この部分は不要だった
query = "restaurant"
candicates = gmaps.places(query, location=latlon, radius=10, language="Japanese")
cr = candicates["results"]
names, place_ids, exact_location = [],[],[]
for i in range(len(cr)):
    if cr[i]['opening_hours']["open_now"] == True:
        names.append(cr[i]["name"])
        place_ids.append(cr[i]["place_id"])
        exact_location.append(cr[i]["geometry"]["location"])
```
